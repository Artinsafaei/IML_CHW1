## Array creation
>Create the followings to numpy array:
* A: a 3\*3 array like this:
$$
A = \left(\begin{array}{cc} 
10 & 20 & 30\\
5 & 10 & 15\\
1 & 5 & 8
\end{array}\right)
$$ 
* B: a 3\*3\*4 that filled with the numbers start from 15 and have steps of 5 to the end. The array should be filled first along the first dimension then the second dimension and at last the third dimension. like this:
$$
B[:, :, 0] = \left(\begin{array}{cc} 
15 & 30 & 45\\
20 & 35 & 50\\
25 & 40 & 55
\end{array}\right)
$$ 

```python
### START CODE HERE ###
A = np.array([[10, 20, 30],
              [ 5, 10, 15],
              [ 1,  5,  8]
              ])

B = np.arange(15, 15 + 5 * (3 * 3 * 4), step=5).reshape((4,3,3))

for i in range(4):
  B[i] = B[i].T
  
### END CODE HERE ###
```

## Actors/Actresses Activity
>For each Actors/Actresses caculate the number of movies in the dataframe, then sort the actors/actresses first by their number of movies in descending order and then by their name in ascending order. 

```python
### START CODE HERE ###
actors_dict = {}

for i in range(len(IMDB_data)):
  film_actors = IMDB_data["Actors"][i].split(',')
  for name in film_actors:
      act = name.strip()
      if act in actors_dict:
        actors_dict[act] += 1
      else:
        actors_dict[act] = 1

result = pd.DataFrame.from_dict({'name': actors_dict.keys(), 'movies':actors_dict.values()}).sort_values(['movies', 'name'], ascending = [False, True])
print(result.to_string(index=False))

### END CODE HERE ###
```

## Plot
>You must plot the number of movies in each year according to the IMDB-Movie-Data.csv file. X-axis is year and y-axis show the number of movies in each year.
> set the x_label, y_label and title for your plot.

```python
### START CODE HERE ###
year_dict = {}
for i in range(IMDB_data['Year'].min(), IMDB_data['Year'].max() + 1):
  year_dict[i] = 0

for y in IMDB_data["Year"]:
    year_dict[y] += 1

plt.rcParams["figure.figsize"] = (12,5)
plt.plot(year_dict.keys(), year_dict.values())
plt.ylabel('Number')
plt.xlabel('Year')
plt.title('Movies in each year')
### END CODE HERE ###
```

## Plot
>You must plot the cos(x) curve. plot 20 cycle of the curve. the frequency of first half of curve is 50 Hz and frequency of second lalf is 200 Hz. 
> set the x_label, y_label and title for your plot.

```python
### START CODE HERE ###
f1 = 50
f2 = 200

t1 = 10/f1
t2 = 10/f2
n = 1000

x1 = np.arange(0, t1, t1/n)
x2 = np.arange(t1, t2+t1, t2/n)

y1 = np.cos(2 * np.pi * f1 * x1)
y2 = np.cos(2 * np.pi * f2 * x2)

x = np.concatenate((x1, x2), axis=0)
y = np.concatenate((y1, y2), axis=0)

plt.rcParams["figure.figsize"] = (20,5)
plt.scatter(x, y)
plt.title('Cosine function')
plt.xlabel('x')
plt.ylabel('y')

### END CODE HERE ###
```
